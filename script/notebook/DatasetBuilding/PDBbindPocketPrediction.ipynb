residue sequence

````python
from Bio import PDB
import warnings

warnings.filterwarnings("ignore")
pre = "../../../data/dta-datasets/PDBbind/PDBbind_file/"
````

````python
# 创建PDB解析器
parser = PDB.PDBParser()
# 读取PDB文件
structure = parser.get_structure("1n4k", pre + "1n4k/1n4k_protein.pdb")
structure
````

````python
# 提取残基序列信息
residues = []
for model in structure:
    for chain in model:
        for residue in chain:
            print(residue.get_id())
            if residue.get_id()[0] != " ":
                # print(residue.get_id()[0])
                continue
            elif PDB.is_aa(residue):
                residues.append(PDB.Polypeptide.three_to_one(residue.get_resname()))
````

````python
# 打印残基序列
sequence = ''.join(residues)
print(sequence)
````

protein residue index

````python
from Bio import PDB
import warnings

warnings.filterwarnings("ignore")
pre = "../../../data/dta-datasets/PDBbind/PDBbind_file/"
````

````python
# 创建PDB解析器
parser = PDB.PDBParser()
# 读取PDB文件
structure = parser.get_structure("1a0t", pre + "1a0t/1a0t_protein.pdb")
structure
````

````python
# 提取残基的索引序号
protein_indices = []
for model in structure:
    for chain in model:
        for residue in chain:
            if PDB.is_aa(residue):
                protein_indices.append(chain.get_id() + "_" + str(residue.get_id()[1]))
````
