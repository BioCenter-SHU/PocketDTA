````python
from Bio import PDB
import warnings

warnings.filterwarnings("ignore")
pre = "../../../data/dta-datasets/PDBbind/PDBbind_file/"
````

````python
# PDB
parser = PDB.PDBParser()
# PDB
structure = parser.get_structure("1n4k", pre + "1n4k/1n4k_protein.pdb")
structure
````

````python
# 
residues = []
for model in structure:
    for chain in model:
        for residue in chain:
            print(residue.get_id())
            if residue.get_id()[0] != " ":
                # print(residue.get_id()[0])
                continue
            elif PDB.is_aa(residue):
                residues.append(PDB.Polypeptide.three_to_one(residue.get_resname()))
````

````python
# 
sequence = ''.join(residues)
print(sequence)
````
