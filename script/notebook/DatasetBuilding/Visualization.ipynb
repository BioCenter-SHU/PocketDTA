# Dataset Distribution visualization

### Filtered Davis

#### Get the Length values of each dataset

```python
import pandas as pd
```

```python
fdavis_df = pd.read_csv('../../../data/dta-datasets/Davis/davis_filtered_datasets.csv')
fdavis_df
```

#### Loop for new DataFrame

```python
length_df = pd.DataFrame(columns=('Drug_Length', 'Target_Length', 'Y'))  
length_df
```

```python
for row in range(fdavis_df.shape[0]):
    smiles = fdavis_df["Drug"][row]
    aa = fdavis_df["Target"][row]
    smile_length = len(smiles)
    aa_length = len(aa) 
    affinity = fdavis_df["Y"][row]
    length_df = length_df.append(pd.DataFrame({
            'Drug_Length':[smile_length],'Target_Length':[aa_length], 'Y':[affinity]}), ignore_index=True)
```

```python
length_df
```

```python
length_df.to_csv('../../../result/csv_output/dataset_length/filtered_davis_length.csv', index=False)
```

### Read for the plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fdavis_df = pd.read_csv('../../../result/csv_output/dataset_length/filtered_davis_length.csv')
fdavis_df
```

```python
davis_df = pd.read_csv('../../../result/csv_output/dataset_length/davis_length.csv')
davis_df
```

```python
kiba_df = pd.read_csv('../../../result/csv_output/dataset_length/kiba_length.csv')
kiba_df
```

```python
pdbbind_df = pd.read_csv('../../../result/csv_output/dataset_length/pdbbind_length.csv')
pdbbind_df
```

```python
fdavis_dict = fdavis_df.to_dict()
fdavis_dict
```

```python
davis_dict = davis_df.to_dict()
davis_dict
```

```python
kiba_dict = kiba_df.to_dict()
kiba_dict
```

```python
pdbbind_dict = pdbbind_df.to_dict()
pdbbind_dict
```

```python
# 创建子图
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(16, 12))  # 创建 3 行 1 列的子图
plt.subplots_adjust(hspace=0.4)
plt.subplots_adjust(wspace=0.4)

# Davis
sns.histplot(data=davis_dict, x="Drug_Length", color="slateblue", bins=20, ax=axes[0, 0])
axes[0, 0].set_xlabel("Length of SMILES string")
axes[0, 0].set_ylabel("Number of drugs")
axes[0, 0].set_title("Davis")
sns.histplot(data=davis_dict, x="Target_Length", color="slateblue", bins=20, ax=axes[1, 0])
axes[1, 0].set_xlabel("Length of protein sequence")
axes[1, 0].set_ylabel("Number of targets")
axes[1, 0].set_title("Davis")
sns.histplot(data=davis_dict, x="Y", color="slateblue", bins=20, ax=axes[2, 0])
axes[2, 0].set_xlabel(r'$pK_d$ values')
axes[2, 0].set_ylabel("Number of DT pairs")
axes[2, 0].set_title("Davis")
# Filtered Davis
sns.histplot(data=fdavis_dict, x="Drug_Length", color="limegreen", bins=20, ax=axes[0, 1])
axes[0, 1].set_xlabel("Length of SMILES string")
axes[0, 1].set_ylabel("Number of drugs")
axes[0, 1].set_title("Filtered Davis")
sns.histplot(data=fdavis_dict, x="Target_Length", color="limegreen", bins=20, ax=axes[1, 1])
axes[1, 1].set_xlabel("Length of protein sequence")
axes[1, 1].set_ylabel("Number of targets")
axes[1, 1].set_title("Filtered Davis")
sns.histplot(data=fdavis_dict, x="Y", color="limegreen", bins=20, ax=axes[2, 1])
axes[2, 1].set_xlabel(r'$pK_d$ values')
axes[2, 1].set_ylabel("Number of DT pairs")
axes[2, 1].set_title("Filtered Davis")
# KIBA
sns.histplot(data=kiba_dict, x="Drug_Length", color="khaki", bins=20, ax=axes[0, 2])
axes[0, 2].set_xlabel("Length of SMILES string")
axes[0, 2].set_ylabel("Number of drugs")
axes[0, 2].set_title("KIBA")
sns.histplot(data=kiba_dict, x="Target_Length", color="khaki", bins=20, ax=axes[1, 2])
axes[1, 2].set_xlabel("Length of protein sequence")
axes[1, 2].set_ylabel("Number of targets")
axes[1, 2].set_title("KIBA")
sns.histplot(data=kiba_dict, x="Y", color="khaki", bins=20, ax=axes[2, 2])
axes[2, 2].set_xlabel('KIBA score')
axes[2, 2].set_ylabel("Number of DT pairs")
axes[2, 2].set_title("KIBA")
# PDBbind
sns.histplot(data=pdbbind_dict, x="Drug_Length", color="lightseagreen", bins=20, ax=axes[0, 3])
axes[0, 3].set_xlabel("Length of SMILES string")
axes[0, 3].set_ylabel("Number of drugs")
axes[0, 3].set_title("PDBbind")
sns.histplot(data=pdbbind_dict, x="Target_Length", color="lightseagreen", bins=20, ax=axes[1, 3])
axes[1, 3].set_xlabel("Length of protein sequence")
axes[1, 3].set_ylabel("Number of targets")
axes[1, 3].set_title("PDBbind")
sns.histplot(data=pdbbind_dict, x="Y", color="lightseagreen", bins=20, ax=axes[2, 3])
axes[2, 3].set_xlabel('affinity values')
axes[2, 3].set_ylabel("Number of DT pairs")
axes[2, 3].set_title("PDBbind")
plt.savefig("../../../result/fig_output/datasets/dataset_length.png", dpi=300, bbox_inches='tight')
```
